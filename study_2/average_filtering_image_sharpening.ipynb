## image sharpening

```python
import cv2 as cv #opencv
import numpy as np
from google.colab.patches import cv_imshow

img=cv.imread('/content/gdrive/MyDrive/MachineVision/sample_img_2/lena.jpg')
blur=cv.blur(img,ksize=(5,5))

detail=np.int32(img)-np.int32(blur)

#a=5
sharpened_img_5=np.int32(img)+5*detail

#a=10 
sharpened_img_10=np.int32(img)+10*detail

#a=20
sharpened_img_20=np.int32(img)+20*detail

cv_imshow(img)
cv_imshow(sharpened_img_5)
cv_imshow(sharpened_img_10)
cv_imshow(sharpened_img_20)
#가중치를 올리며 비교
```
