```python
def addsaltandpeppernoise (image, ratio) :
  row,col,ch = image.shape
  out = np.copy(image)
  #salt(흰 점)
  num_salt = np.ceil(image.size*ratio)
  coords = [np.random.randint(0, i-1, int(num_salt)) for i in image.shape]
  out[coords] = 255
  
  #pepper(검은 점)
  num_pepper = np.ceil(image.size*ratio)
  coords = [np.random.randint(0, i-1, int(num_pepper)) for i in image.shape]
  out[coords] = 0
 
  return out
```

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv_imshow
img=cv.imread('/content/gdrive/MyDrive/MachineVision/paca.jpg')
blur =cv.blur(img,ksize=(10,10))
cv_imshow(img) #원래 사진
cv_imshow(blur) #흐려진 알파카


detail=np.int32(img)-np.int32(blur)
#a=20
sharpened_img_20=np.int32(img)+2*detail
cv_imshow(sharpened_img_20) #선명한 알파카


img = addsaltandpeppernoise(img, ratio=0.01)
median_img_3 =cv.medianBlur(img, ksize=15) #노이즈 제거
cv_imshow(median_img_3)
```
