## Gaussian filter

```python
import cv2 as cv #opencv
import numpy as np
from google.colab.patches import cv_imshow

img=cv.imread('/content/gdrive/MyDrive/MachineVision/sample_img_2/opencv-logo-white.png')
blur1=cv.GaussianBlur(img,ksize=(7,7),sigmaX=3)
blur2=cv.GaussianBlur(img,ksize=(11,11),sigmaX=5)

cv_imshow(img)
cv_imshow(blur1)
cv_imshow(blur2)
```

## Median filter

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv_imshow
def addsaltandpeppernoise (image, ratio) :
  row,col,ch = image.shape
  out = np.copy(image)
  #salt( )
  num_salt = np.ceil(image.size*ratio)
  coords = [np.random.randint(0, i-1, int(num_salt)) for i in image.shape]
  out[coords] = 255
  
  #pepper( )
  num_pepper = np.ceil(image.size*ratio)
  coords = [np.random.randint(0, i-1, int(num_pepper)) for i in image.shape]
  out[coords] = 0
 
  return out
```

```python
img=cv.imread("/content/gdrive/MyDrive/MachineVision/sample_img_2/opencv-logo-white.png")

img = addsaltandpeppernoise(img, ratio=0.01)
median_img_3 =cv.medianBlur(img,ksize=3) #  ksize 1  
median_img_10 =cv.medianBlur(img,ksize=11)
median_img_1 =cv.medianBlur(img,ksize=1)

cv_imshow(img)
cv_imshow(median_img_3)
cv_imshow(median_img_10)
cv_imshow(median_img_1)
```
